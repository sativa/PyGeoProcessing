```python
#import gdal libraries 
from osgeo import gdal
from osgeo import _gdal
from osgeo import gdalnumeric
from osgeo.gdal_array import *
from osgeo.gdalconst import *
import time, numpy, os, ogr
```

```python
# utility to read blocks 
import imp
utils2 = imp.load_source('utils2', 'H:/utils.py')
#Edit the path for the Landsat TOA Reflectance data directory 
dirpath='H:/landsat'
```

```python
def Cloud_Function(blue, green, red, nir,swir1, swir2, thermal):
    # Start time for then time profiling of the algorithm 
    startTime = time.time()

    #Get Raster file format driver 
    driver = gdal.GetDriverByName( 'GTiff' ) 
     # open file 
    tmp = gdal.Open(blue,GA_ReadOnly)         
    #Fetches the coefficients for transforming between pixel/line (P,L) raster space, and projection coordinates (Xp,Yp) space.
    geoT = tmp.GetGeoTransform() 
    # Read Projection           
    proJ = tmp.GetProjection()              
    
    # Read Number of Column and Rows                
    cols = tmp.RasterXSize                  
    rows = tmp.RasterYSize                
    # Read number of bands in image 
    bands = tmp.RasterCount                 
    BLUEtmp = tmp.GetRasterBand(1)
    
    # Read Red Band 
    redF = gdal.Open(red,GA_ReadOnly)
    REDtmp = redF.GetRasterBand(1)
    # Read Green Band 
    greenF = gdal.Open(green,GA_ReadOnly)
    GREENtmp = greenF.GetRasterBand(1)
    # Read NIR Band 
    nirF = gdal.Open(nir,GA_ReadOnly)
    NIRtmp = nirF.GetRasterBand(1)
    # Read SWIR1 Band 
    swir1F = gdal.Open(swir1,GA_ReadOnly)
    SWIR1tmp = swir1F.GetRasterBand(1)
    # Read SWIR2 Band 
    swir2F = gdal.Open(swir2,GA_ReadOnly)
    SWIR2tmp = swir2F.GetRasterBand(1)
    # Read Thermal Band 
    ThermalF = gdal.Open(thermal,GA_ReadOnly)
    THER1tmp = ThermalF.GetRasterBand(1)
         
    Output_filename='Output/CloudScore.tif'   
    OutputDS = driver.Create(Output_filename, cols, rows, 1, GDT_Float32)
    OutputDS.SetGeoTransform(tmp.GetGeoTransform())
    OutputDS.SetProjection(tmp.GetProjection())
    outBand_data = OutputDS.GetRasterBand(1)
    
    mask_filename='Output/Cloudmask.tif'   
    OutputDS2 = driver.Create(mask_filename, cols, rows, 1, GDT_Float32)
    OutputDS2.SetGeoTransform(tmp.GetGeoTransform())
    OutputDS2.SetProjection(tmp.GetProjection())
    maskBand_data = OutputDS2.GetRasterBand(1)
    
    # Blockwise optmization to speedup the image processing 
    # block size for the landsat data processing
    yBlockSize=2000
    xBlockSize=2000

    for i in range(0, rows, yBlockSize):
        if i + yBlockSize < rows:
            numRows = yBlockSize
        else:
            numRows = rows - i
        for j in range(0, cols, xBlockSize):
            if j + xBlockSize < cols:
                numCols = xBlockSize
            else:
                numCols = cols - j
                
            REDdata = BandReadAsArray(REDtmp,j,i,numCols, numRows).astype(numpy.float) 
            GREENdata = BandReadAsArray(GREENtmp,j,i,numCols, numRows).astype(numpy.float)                
            BLUEdata = BandReadAsArray(BLUEtmp,j,i,numCols, numRows).astype(numpy.float)   
            NIRdata = BandReadAsArray(NIRtmp,j,i,numCols, numRows).astype(numpy.float)     
            SWIR1data = BandReadAsArray(SWIR1tmp,j,i,numCols, numRows).astype(numpy.float) 
            SWIR2data = BandReadAsArray(SWIR2tmp,j,i,numCols, numRows).astype(numpy.float) 
            THERMALdata = BandReadAsArray(THER1tmp,j,i,numCols, numRows).astype(numpy.float)                 
   
            # band threshold based on the blue band 
            CloudScore=numpy.minimum((BLUEdata-0.1)/(0.3-0.1),1)
            #Clouds are bright in all visible bands. We added the all RGB bands and give the threshold of 0.2 and 0.8
            CloudScore=numpy.minimum(((REDdata+GREENdata+BLUEdata-0.2)/(0.8-0.2)),CloudScore)
            #Clouds are bright in infrared and both shortwave bands.  We added the all RGB bands and give the threshold of 0.3 and 0.8
            CloudScore=numpy.minimum(((NIRdata+SWIR1data+SWIR2data-0.3)/(0.8-0.3)),CloudScore)            
            #Shadow masking 
            Shadow_Mask=(NIRdata-1)/(-0.1-1)    
            CloudScore=numpy.where(Shadow_Mask<0.1,Shadow_Mask, CloudScore)
            # Precipitable water (Cloud)  and cloud shadows are reasonably cool in temperature than soil and water body.
            CloudScore=numpy.minimum(((THERMALdata-305)/(285-305)),CloudScore)
            # Normalized Difference Snow Index (NDSI) for snow and water 
            NDSI=(GREENdata-SWIR1data)/(GREENdata+SWIR1data)
            
            CloudScore=numpy.minimum(((NDSI-0.8)/(0.6-0.8)),CloudScore)
            # Final Cloud Scoring System 
            CloudScore=CloudScore*100
            mask=numpy.where(CloudScore>1,1,0)
            
            # Write output
            outBand_data.WriteArray(CloudScore, j, i)
            maskBand_data.WriteArray(mask, j, i)
            # delete temporary data to release memory 
            del REDdata, NIRdata, CloudScore, mask, NDSI, Shadow_Mask
    
   

    # create output shapeffile name
    output_shp = "Output/CloudMask_ShapeFile"
    shp_driver = ogr.GetDriverByName("ESRI Shapefile")
    output_shapefile = shp_driver.CreateDataSource( output_shp + ".shp" )
    new_shapefile = output_shapefile.CreateLayer(output_shp, srs = None )
    # Specify the connectivity  to the gdal.Polygonize according to the accuracy (takes 70% of the total time)
    gdal.Polygonize(maskBand_data, None, new_shapefile, 10, [], callback=None)
   
    endTime = time.time()
    print 'The cloud masking process took ' + str(endTime - startTime) + ' seconds'
                
```

```python
if __name__ == "__main__":
    os.chdir(dirpath)
    #Read all required bands from the directory 
    for filename in os.listdir(dirpath):  
            name, extension = os.path.splitext(filename)
            if extension.lower() == ".tif":
                if 'B2' in filename:                    
                    blue= filename
                if 'B3' in filename:   
                    green= filename
                if 'B4' in filename:  
                    red= filename
                if 'B5' in filename: 
                    nir= filename
                if 'B6' in filename:
                    swir1= filename
                if 'B7' in filename: 
                    swir2= filename
                if 'B10' in filename:
                    thermal= filename
                
    #Call the cloud function 
    Cloud_Function(blue, green, red, nir,swir1, swir2, thermal)
```
