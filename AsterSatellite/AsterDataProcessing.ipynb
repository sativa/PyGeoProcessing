```python
#Import Hierarchical Data Format (HDF) Images and Make List                 
def import_hdfimages(dirpath):
    print(' import_images')
    Aster = list()
    for filename in os.listdir(dirpath):
            name, extension = os.path.splitext(filename)
            if extension.lower() == ".hdf":
                Aster.append((filename))      
    #print(' end import_images')
    print Aster
    return Aster
# Digital Number to Radiance  
def DnToRad ( data, band, gain ):
    gain_settings={'high':0, 'normal':1, 'low1':2, 'low2':3}
    conversion_factors=numpy.array([[0.676, 1.688, 2.25, 0.0],\
                                    [0.708, 1.415, 1.89, 0.0],\
                                    [0.423, 0.862, 1.15, 0.0],\
                                    [0.423, 0.862, 1.15, 0.0],\
                                    [0.1087, 0.2174, 0.2900, 0.2900],\
                                    [0.0348, 0.0696, 0.0925, 0.4090],\
                                    [0.0313, 0.0625, 0.0830, 0.3900],\
                                    [0.0299, 0.0597, 0.0795, 0.3320],\
                                    [0.0209, 0.0417, 0.0556, 0.2450],\
                                    [0.0159, 0.0318, 0.0424, 0.2650]])
    radiance = (data-1.)*conversion_factors[band-1][gain_settings[gain]]
    return radiance
#Radiance to Refelecatnce 
    
def RadToRefl ( radiance, band, day, sza, solar_constant_scheme="ThomeEtAl2" ):
    schemes={'Smith':0, 'ThomeEtAl1':1, 'ThomeEtAl2':2}
    solar_constant=numpy.array([[1845.99, 1847, 1848],    \
                                [1555.74, 1553, 1549],    \
                                [1119.47, 1118, 1114],    \
                                [231.25, 232.5,  225.4 ], \
                                [79.81,  80.32,  86.63 ], \
                                [74.99,  74.92,  81.85 ], \
                                [68.66,  69.20,  74.85 ], \
                                [59.74,  59.82,  66.49 ], \
                                [56.92,  57.32,  59.85]]) #B9
    d = 1.0 - 0.01672*numpy.cos(numpy.radians(0.9856*(day-4)))
    toaRefl = (numpy.pi*radiance*d*d)/(solar_constant[band-1][schemes[solar_constant_scheme]]* numpy.cos ( numpy.radians (sza) ))
    return toaRefl
                    
```

```python
# Refer Aster Data Processing Manual from NASA
if __name__ == "__main__":
    #dirpath='/home/hydroinfo/Desktop/Aster/Ast_2008-2011'
    #dirpath='C:/Users/Home/Desktop/aster'
    dirpath='/home/hydroinfo/Desktop/Aster/Ast_2008-2011/output'
    
    bandlist1 = ['VNIR_Swath:ImageData1','VNIR_Swath:ImageData2','VNIR_Swath:ImageData3N','VNIR_Swath:ImageData3B','SWIR_Swath:ImageData4','SWIR_Swath:ImageData5','SWIR_Swath:ImageData6','SWIR_Swath:ImageData7','SWIR_Swath:ImageData8','SWIR_Swath:ImageData9','TIR_Swath:ImageData10','TIR_Swath:ImageData11','TIR_Swath:ImageData12','TIR_Swath:ImageData13','TIR_Swath:ImageData14']
    bandlist = ['VNIR_Swath:ImageData3N']
   
    for filename in os.listdir(dirpath):
         name, extension = os.path.splitext(filename)
         if extension.lower() == ".hdf":        
                for band in bandlist:
                    startTime = time.time()
                    dataset = band
                    srcfile = "HDF4_EOS:EOS_SWATH:%s:%s" % (filename, dataset)
                    name, extension = os.path.splitext(filename)
                    output_filename=str(name)+'_'+'VNIR_Swath:ImageData2.tif'
                    output_filename2=str(name)+'_'+'VNIR_Swath:ImageData3N.tif'
                    xml_filename=str(filename)+'.xml'
                    
                    dom = parse(xml_filename)
                    date1 = dom.getElementsByTagName('CalendarDate')
                    date2=date1[0].firstChild.nodeValue
                   time1 = dom.getElementsByTagName('TimeofDay')
                    time2=time1[0].firstChild.nodeValue

                    library = xml.dom.minidom.parse (xml_filename)
                    tracks = library.documentElement.getElementsByTagName ( 'PSAName' )
                    tracks2 = library.documentElement.getElementsByTagName ( 'PSAValue' )


                    metadata = list()
                    for track, track2 in zip(tracks,tracks2):
                        metadata.append((track.childNodes [ 0 ].nodeValue  + str(':') + track2.childNodes [ 0 ].nodeValue))
                    for line in metadata:
                        if 'Solar_Azimuth_Angle' in line:
                            Solar_Azimuth_Angle= str(line.strip().split(':')[1].strip() )
                            Solar_Azimuth_Angle =float(Solar_Azimuth_Angle)   
                        if 'Solar_Elevation_Angle' in line:
                            Solar_Elevation_Angle= str(line.strip().split(':')[1].strip() )
                            Solar_Elevation_Angle =float(Solar_Elevation_Angle)   
                    #sys.exit()
                    
                    date = datetime.datetime.strptime(date2, '%Y-%m-%d')
                    tt=date.timetuple()
                    doy=int(tt[7])
                    year=int(tt[0])
                    
                    sun_elevation = float(Solar_Elevation_Angle)
                    constant=(numpy.cos((90-sun_elevation) * numpy.pi/180 ) / (numpy.pi * (1+0.01672 * numpy.sin(2 * numpy.pi * (doy - 93.5) / 365)) ** 2))
                    solarzenithangle=((90-sun_elevation) * numpy.pi/180 )
                    print 'Processing data for----' + str( date)

                    
                    driver = gdal.GetDriverByName( 'GTiff' )
                    
                    tmp1 = gdal.Open((output_filename),GA_ReadOnly)    
                    btmp1 = tmp1.GetRasterBand(1) 
                    tmp2 = gdal.Open(output_filename2,GA_ReadOnly)
                    btmp2 = tmp2.GetRasterBand(1)   
             
                    raw_data = tmp2.ReadAsArray()
                    geoT = tmp2.GetGeoTransform()
                    proJ = tmp2.GetProjection()
                    rows = tmp2.RasterYSize
                    cols = tmp2.RasterXSize
                    yBlockSize = 1000
                    xBlockSize = 1000
                    

                    #NDVI 
                    output_filename=str(name)+'_NDVI'+'.tif'   
                    ndviDS = driver.Create(output_filename, cols, rows, 1, GDT_Float32)
                    ndviDS.SetGeoTransform(tmp2.GetGeoTransform())
                    ndviDS.SetProjection(tmp2.GetProjection())
                    outBand_ndvi = ndviDS.GetRasterBand(1)
                    #SAVI    
                    output_filename2=str(name)+'_SAVI'+'.tif'   
                    saviDS = driver.Create(output_filename2, cols, rows, 1, GDT_Float32)
                    saviDS.SetGeoTransform(tmp2.GetGeoTransform())
                    saviDS.SetProjection(tmp2.GetProjection())
                    outBand_savi = saviDS.GetRasterBand(1)
                   
                    #LAI
                    output_filename4=str(name)+'_LAI'+'.tif'   
                    laiDS = driver.Create(output_filename4, cols, rows, 1, GDT_Float32)
                    laiDS.SetGeoTransform(tmp2.GetGeoTransform())
                    laiDS.SetProjection(tmp2.GetProjection())
                    outBand_lai = laiDS.GetRasterBand(1)  
                    

                    for i in range(0, rows, yBlockSize):
                        if i + yBlockSize < rows:
                            numRows = yBlockSize
                        else:
                            numRows = rows - i
                        for j in range(0, cols, xBlockSize):
                            if j + xBlockSize < cols:
                                numCols = xBlockSize
                            else:
                                numCols = cols - j
                            data1 = BandReadAsArray(btmp1,j,i,numCols, numRows).astype(numpy.float)
                            data2 = RadToRefl ( DnToRad ( data1, 2, 'high'), 2, doy, solarzenithangle)
                            data4 = BandReadAsArray(btmp2,j,i,numCols, numRows).astype(numpy.float)
                            data3 = RadToRefl ( DnToRad ( data4, 3, 'normal'), 3, doy, solarzenithangle)
                            
                            mask = numpy.greater(data3 + data2, 0)
                            ndvi1 = numpy.choose(mask, (0, (data3 - data2) / (data3 + data2)))
                            mask_ndvigtpt0=numpy.greater_equal(ndvi1, 0)
                            ndvi= numpy.choose(mask_ndvigtpt0, (0, (ndvi1)))
                            savi = numpy.choose(mask, (0, ((1+0.1)*(data3-data2)/(1*(data3+data2+0.1)))))

                            mask_saviltpt817=numpy.less(savi, 0.817)
                            mask_savigtpt817=numpy.greater_equal(savi, 0.817)            
                            
                            lai1= numpy.choose(mask_saviltpt817, (0, (11*savi*savi*savi)))
                            lai2= numpy.choose(mask_savigtpt817, (0, (6)))
                            lai=numpy.choose(mask, (0, (lai1+lai2)))  
                            
                            outBand_ndvi.WriteArray(ndvi, j, i)
                            outBand_savi.WriteArray(savi, j, i)
                            outBand_lai.WriteArray(lai, j, i)
                savefile1=str(year)+'_'+str(doy)+'ndvi.png'
                fig = pylab.imshow ( ndvi )
                pylab.title('NDVI for-Date-- %s----DOY--%s--Time--%s '%(date, doy, time2))	
                pylab.colorbar(fig, orientation='horizontal', shrink=0.8)
                pylab.savefig(savefile1, transparent=True)
                pylab.close()

                endTime = time.time()
                print 'The VI Processing for'+ str(date) +'took--------'+ str(endTime - startTime) + ' seconds'
                print '_______________________________________________________________________________________'
                print(' Done')
                gdal.SetConfigOption('HFA_USE_RRD', 'YES') 
                del data1, data2, data4, data3, ndviDS, saviDS, laiDS 
```
